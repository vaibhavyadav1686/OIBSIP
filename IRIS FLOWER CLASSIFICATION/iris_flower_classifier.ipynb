```python
import pandas as pd
import seaborn as sns
sns.set_palette('husl')
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
dataset = pd.read_csv('Iris.csv')
print(dataset.head()) 
print(dataset.info())
X = dataset.drop(['Id','Species'], axis = 1)
y = dataset['Species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 5)
```

```python
#Logistic Regression
model1 = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg').fit(X_train, y_train)
prediction = model1.predict(X_test)
print("LOGISTIC REGRESSION: ")
print(f'Test Accuracy: {accuracy_score(y_test, prediction)}')
print(f'Classification Report: \n {classification_report(y_test, prediction)}')
```

```python
#Linear Discriminant Analysis
model2 = LinearDiscriminantAnalysis().fit(X_train, y_train)
prediction = model2.predict(X_test)
print("LINEAR DISCRIMINANT ANALYSIS: ")
print(f'Test Accuracy: {accuracy_score(y_test, prediction)}')
print(f'Classification Report: \n {classification_report(y_test, prediction)}')
```

```python
#K Nearest Neighbours
model3 = KNeighborsClassifier().fit(X_train, y_train)
prediction = model3.predict(X_test)
print("K NEAREST NEIGHBOURS: ")
print(f'Test Accuracy: {accuracy_score(y_test, prediction)}')
print(f'Classification Report: \n {classification_report(y_test, prediction)}')
```

```python
#DECISION TREE CLASSIFIER
model4 = DecisionTreeClassifier().fit(X_train, y_train)
prediction = model4.predict(X_test)
print("DECISION TREE CLASSIFIER: ")
print(f'Test Accuracy: {accuracy_score(y_test, prediction)}')
print(f'Classification Report: \n {classification_report(y_test, prediction)}')
```

```python
#GAUSSIAN NAIVE BAYES
model5 = GaussianNB().fit(X_train, y_train)
prediction = model5.predict(X_test)
print("GAUSSIAN NAIVE BAYES CLASSIFIER: ")
print(f'Test Accuracy: {accuracy_score(y_test, prediction)}')
print(f'Classification Report: \n {classification_report(y_test, prediction)}')
```

```python
#SUPPORT VECTOR MACHINE
model6 = SVC(gamma='auto').fit(X_train, y_train)
prediction = model6.predict(X_test)
print("SUPPORT VECTOR MACHINE: ")
print(f'Test Accuracy: {accuracy_score(y_test, prediction)}')
print(f'Classification Report: \n {classification_report(y_test, prediction)}')
```
